3. Write a Python program to find the repeated items of the following tuple. 

`t1 = (1,1,3,1,3,3,2,1,4,5)`

````python
t1 = (1,1,3,1,3,3,2,1,4,5)
d1=dict()
for i in t1:
    if i in d1:
        d1[i]+=1
    else:
        d1[i]=1
lst = [x for x,y in d1.items() if y>1]
lst
````
