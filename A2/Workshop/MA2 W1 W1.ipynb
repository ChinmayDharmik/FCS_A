1. What is the difference between append and extend? Write two example codes to illustrate the difference.

````python
'''
append -- > append() adds a single element to the end of the list.
extend -- > xtend() can add multiple individual elements to the end of the list.
'''
lst1 = ['x',1 , 2 , 3]
lst2 = [4,5]
lst1.append(lst2)
print(lst1)
lst1.extend(lst2)
print(lst1)
````

2. What does `del` do? Write an example to illustrate this.

````python
my_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# deleting the third item
del my_list[2]

# Output: [1, 2, 4, 5, 6, 7, 8, 9]
print(my_list)

# deleting items from 2nd to 4th
del my_list[1:4]

# Output: [1, 6, 7, 8, 9]
print(my_list)

# deleting all elements
del my_list[:]

# Output: []
print(my_list)
````

3. What is the difference between `remove` and `pop`?
