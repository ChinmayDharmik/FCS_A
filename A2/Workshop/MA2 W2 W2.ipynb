2. Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.

````python
def fact(num):
    if num >= 0:
        mul=1
        for i in range(1,num+1):
            mul*=i
        return mul
    return -1
fact(-4)
````

3. Write a Python function that accepts a string and calculates the number of upper case letters and lower case letters.
Sample String: 'The quick Brow Fox'
Expected Output :
No. of Upper case characters : 3
No. of Lower case Characters: 12

````python
def CharDist(str1):
    up=0
    lo=0
    for  i in str1:
        if i.isupper():
            up+=1
        if i.islower():
            lo+=1
    return up, lo
CharDist('The quick Brow Fox')
````

4. Write a function that merges two lists. This function must have 3 parameters: `list_1`,` list_2`, and `mode`. The function must return a new list that results from merging the input lists: every other element of the output starting from the first should be taken from the `list_1`, while every other element starting from the second should be taken from the `list1_2`.

* Example 1: If list_1 = [1, 2, 3, 4] and list_2 = [-1, -2, -3, -4] the output should be [1, 2, 3, 4, -1, -2, -3, -4 ]
* Example 2: If list_1 = ['a', 'b', 'c'] and list_2 = [1, 2, 3] the output should be ['a', 'b', 'c', 1, 2, 3]
You can assume that the input lists are not empty and have the same length.

By default, `mode` should be equal to "show". If mode equals "show", the output should be as described above. If mode equals "hide", all elements in the output list should be hidden, i.e., replaced with '?'.

Example 1: If mode = 'hide', list_1 = [1, 2, 3, 4] and list_2 = [-1, -2, -3, -4] the output should be ['?', '?', '?', '?', '?', '?', '?', '?'] with 8 question marks.
Example 2: If mode = 'hide', list_1 = ['a', 'b', 'c'] and list_2 = [1, 2, 3] the output should be ['?', '?', '?', '?', '?', '?'] with six question marks.

````python
def pf4(list_1,list_2,mode='show'):
    if mode =='show':
        print(list_1+list_2)
    elif mode == 'hide':
        out = []
        for i in range(len(list_1+list_2)):
            out.append('?')
        print(out)

pf4([1,2,3,4],[1,2,3,4,5],mode='hide')

````
