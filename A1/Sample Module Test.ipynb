Imagine your program crawls some text from the web, but the result contains some weird characters. 
```
txt = "   \n\n   \t Th|s |S 50m3 c0D3,,, \n \t "
```

Translate the above-mentioned message into plain English (as shown below) using string operations.
```
this is some code.
```

````python
txt = "   \n\n   \t Th|s |S 50m3 c0D3,,, \n \t "
print(txt.replace('\n','').replace('5','s').replace('0','o').replace('3','e').replace('\t','').replace('|','i').replace(',','').strip().lower())
````

Extend the following program that simulates the behaviour of a simple calculator to meet the two requirements mentioned below.
```python
Task 1: Let the user pick the arithmetic operation they would like to perform.
Task 2: Let the user repeatedly perform the arithmetic operations, until the user decides to exit.

Mentioned below is a sample output of the Python program that you need to develop in this question.
```

````python
cont = True
while cont == True:
    print('''Select the arithemetic operation.
    1. addition
    2. subtraction
    3. multiplication
    4. division
    5. exponentiation
    6. integer_division''')
    choice = int(input('Enter your choice (1/2/3/4/5/6):'))

    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    if choice==1:
        addition = num1 + num2
        print(f"Addition of {num1} + {num2} = {addition}")
    elif choice ==2:
        subtraction = num1 - num2
        print(f"Subtraction of {num1} - {num2} = {subtraction}")
    elif choice ==3:
        multiplication = num1 * num2
        print(f"Multiplication of {num1} x {num2} = {multiplication}")
    elif choice ==4:
        division = num1 / num2
        print(f"Division of {num1} / {num2} = {division}")
    elif choice ==5:
        exponentiation = num1 ** num2
        print(f"Exponentiation of {num1} ** {num2} = {exponentiation}")
    elif choice ==6:
        integer_division = num1 // num2
        print(f"Integer division of {num1} // {num2} = {integer_division}")
    else:
        print('invalid option')

    cont = input("Let's do next calculation? (y/n): ").lower().startswith('y')
````

Write a Python program that computes the minimum number of bills and coins needed to make change for each transaction in a store. This program is used by a cashier and it should keep on asking for the amount of change for each transaction from the cashier as an input until the cashier's shift is over.

For this problem, assume that we only have the following bills and coins.

Bills: $20, $10 and $1
Coins: 50 cents ($0.5), 20 cents ($0.2) and 10 cents ($0.1)
For instance, if your amount of change is $91.93 in change, you would need the following bills and coins.

Bills: four $20, one $10, and one $1.
Coins: one 50 cents and two 10 cents.
Use the intermediate variables bills_20, bills_10, bills_1, coins_50, coins_20 and coins_10 to store the number of bills and coins needed for each amount specified. The output of the minimum number of bills and coins should be displayed in the following format (for the above-described example of $91.93).

4 twenty bills

1 one bill

2 twenty coins
Use conditional statements to handle the singular and plural version of the above output. When only one bill/coin is needed, you should use the singular version of the word. When more than one bill/coin is needed, the plural of the label should be used. Mentioned below are the singular and plural versions of bills and coins.

singular	plural

Let the cashier enter the amount to change (e.g., $91.93) as an input to the program.

This process should be repeatedly performed until the cashier's shift is over. Therefore, in each iteration, your program has to ask the cashier if the shift is over. If the shift is not over, keep on asking the amount of change until the cashier specifies that the shift is over.

Mentioned below are two sample outputs of the Python program that you need to develop in this question.

````python
bills_20, bills_10, bills_1, coins_50, coins_20 , coins_10 = 0,0,0,0,0,0
cont = input('Is your shift over (y/n)?')
while cont == 'n':
    total = float(input('Mention the amount of change:'))
    bills  = total // 1
    coins = 100*(total % 1)

    bills_20 = int(bills//20)
    bills = bills%20
    bills_10 = int(bills//10)
    bills = bills%10
    bills_1 = int(bills)

    coins_50 = int(coins//50)
    coins= coins%50
    coins_20 = int(coins//20)
    coins=coins%20
    coins_10 = int(coins//10)

    output = ''
    if bills_20>=1:
        output += f'{bills_20} twenty bill\n' if bills_20==1 else f'{bills_20} twenty bills\n'
    if bills_10>=1:
        output += f'{bills_10} ten bill\n' if bills_10==1 else f'{bills_10} ten bills\n'
    if bills_1>=1:
        output += f'{bills_1} one bill\n' if bills_1==1 else f'{bills_1} one bills\n'
    if coins_50>=1:
        output += f'{coins_50} fifty coin\n' if coins_50==1 else f'{coins_50} fifty coins\n'
    if coins_20>=1:
        output += f'{coins_20} twenty coin\n' if coins_20==1 else f'{coins_20} twenty coins\n'
    if coins_10>=1:
        output += f'{coins_10} ten coin\n' if coins_10==1 else f'{coins_10} ten coins\n'
    print(output)
    cont = input('Is your shift over (y/n)?')   
````
