## Question 5&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(40 Points)

```python
# Write your solution here
print("Where would you like to place your toy on the integer 2D coordinate grid?")
x = int(input("Enter x: "))
y = int(input("Enter y: "))
cont = True
while cont:
    direction = input("Enter direction :")
    if direction.lower() == 'up':
        y+=1
        print(f'New location: x = {x}, y={y}')
    elif direction.lower() == 'down':
        y-=1
        print(f'New location: x = {x}, y={y}')
    elif direction.lower() == 'left':
        x-=1
        print(f'New location: x = {x}, y={y}')
    elif direction.lower()=='right':
        x+=1
        print(f'New location: x = {x}, y={y}')
    else:
        print("unsupported operation")
    
    print("-------------------------------")
    cont = input("Would you like to play again (y/n)?").lower()=='y'
print("See you next time!")
```
