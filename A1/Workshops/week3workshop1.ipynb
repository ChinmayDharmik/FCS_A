4. Create simulated blood glucose readings for 500 persons without diabetes. The normal values are below 126 mg/dL and over 60 mg/dL. Write Python code that uses the random library to generate and print 500 values between these ranges.

5. In Australia, blood glucose levels are reported in mmol/L instead of mg/dL. The conversion is:
mmol/L = (mg/dL) / 18
Write Python code that first prompts the user to specify the preferred unit: (mmol/L or mg/dL) and then generates 500 data points in the specified units.  

````python
import random
count = 1
choice  = input("mmol or mgdl: ")

while count <=500 :
    a=random.randint(60,126)
    if choice == "mmol":
        a = a/18
    print(a)
    count = count + 1
````

7. Write a Python program to print n natural numbers in descending order using a while loop.

````python
num = int(input("Enter a range: "))
result= "Result:"
while num >=1 :
    result += " "+str(num)
    num-=1
print(result)
````

1. Print the following pattern using while loops.
```
*
* *
* * *
* * * *
```

````python
i = 1 
while i<=4:
    j=1
    str1 = ''
    while j<=i:
        str1 += "* "
        j+=1
    i+=1
    print(str1)
````

4. Take 10 integers from the keyboard using while loops and print their average value and sum on the screen.

````python
counter = 10
summation = 0 
while counter >=1:
    num = int(input("enter a number :"))
    summation +=num
    counter -=1
average = summation/10
summation, average
````
