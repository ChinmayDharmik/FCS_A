3. Write a Python code to check whether the user entered year is a leap year or not (Leap year definition: exactly divisible by 4 except for century years (years ending with 00). The century year is a leap year only if it is perfectly divisible by 400).

```python
userInput=int(input("enter the year"))
if userInput%100 == 0:
    if userInput%400 == 0:
        print("the century year is a leap year")
    else:
        print("the century year is not a leap year")
elif userInput%4 == 0:
    print("the year is a leap year")
else:
    print("Not a leap year")
```
