**Hint: A complex number is represented as â€œ x + yi â€œ**

Even though we have a library `cmath` in python, for this question lets leverage the power of operator overloading to create a complex number class and perform various mathemaical operations. You have been provided with a stub.

For this question you are expected to create a class `ComplexNumber` and overload the metohds for performing -
1. Addition of two complex numbers

```(a + bi) + (c + di) = (a+c) + (b+d)i```

2. Subtraction of two complex numbers

```(a + bi) - (c + di) = (a-c) + (b-d)i```

3. Multiplication of two complex numbers

```(a + bi) * (c + di) = (acâˆ’bd) + (ad+bc)i``` ***(i^2 = -1 Remember?)***

4. Division of two complex numbers

```(a + bi) / (c + di) = ([(acâ€‰+â€‰bd)â€‰+â€‰(bcâ€‰âˆ’â€‰ad)i]/(c^2â€‰+â€‰d^2)```

Example Call -
```

Expected Output

```

````python
class ComplexNumber:
    def __init__(self, real, imag):
        self.re = real
        self.im = imag
    
    #write code here
    def __add__(self,other):
        return str(ComplexNumber(self.re + other.re,self.im + other.im))
    
    def __sub__(self,other):
        return str(ComplexNumber(self.re - other.re,self.im - other.im))

    def __mul__(self,other):
        real = self.re*other.re - self.im*other.im
        imag = self.im*other.re + self.re*other.im
        return str(ComplexNumber(real,imag))
        
    def __truediv__(self, other):
        real = (self.re*other.re + self.im*other.im)/(other.re**2+other.im**2)
        imag = (self.im*other.re - self.re*other.im)/(other.re**2+other.im**2)
        return str(ComplexNumber(real,imag))
        
# DO NOT MODIFY CODE BELOW THIS LINE
    def __str__(self):
        if self.im == 0:
            return '%.2f' % self.re
        if self.re == 0:
            return '%.2fi' % self.im
        if self.im < 0:
            return '%.2f - %.2fi' % (self.re, -self.im)
        else:
            return '%.2f + %.2fi' % (self.re, self.im)

def complexCalculator(real1 , imaginary1 , real2, imaginary2):
    c1 = ComplexNumber(real1,imaginary1)
    c2 = ComplexNumber(real2,imaginary2)
    return [ str(c1+c2) , str(c1-c2) , str(c1*c2) , str(c1/c2) ]

c1 = ComplexNumber(1, 2)
c2 = ComplexNumber(3, -4)
print(c1 + c2)
print(c1 - c2)
print(c1 * c2)
print(c1 / c2)
````
