```python
## REFECTIVE LEARNING QUESTION 2
import math 
class Circle:
    def __init__(self,r):
        self.radius=r
    def Area(self):
        return math.pi*self.radius**2
    def Perimeter(self):
        return 2*math.pi*self.radius
cir = Circle(5)
cir.Area(),cir.Perimeter()
```

```python
## Practice Inheritance Part 1 Question 2
class Person:
    name = ''
    age = 0
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def display(self):
        print(f'Person Name : {self.name}\nPerson Age : {self.age}')


class Student(Person):
    section = ''
    def __init__(self,name,age,Section = 1):
        super().__init__(name,age)
        self.section = Section
    def displayStudent(self):
        print(f'Student Name : {self.name}\nStudent Age : {self.age}\nStudent Section : {self.section}')

# Testing Person class
P = Person("Tomas Wild", 37)
P.display()
print()
S = Student("Albert", 23 , "Mathematics")
S.displayStudent()
```

```python
## Practice Inheritance Part 2
class BankAccount: 
    """Bank Account protected by a pin number."""
    _pin = 'pin'
    _balance = 0
    def __init__(self, pin='pin'): 
        """Initial account balance is 0 and pin is 'pin'."""
        self._balance = 0
        self._pin=pin
    def deposit(self, pin, amount): 
        """Increment account balance by amount and return new balance."""
        if self._check_pin(pin):
            self._balance+=amount
        else:
            print("Invalid Pin")

    def withdraw(self, pin, amount): 
        """Decrement account balance by amount and return amount withdrawn."""
        if self._check_pin(pin):
            if self._balance >= amount:
                self._balance-= amount
            else:
                print("Insufficient Balance")
        else:
            print("Invalid Pin")
    def _check_pin(self,pin):
        return self._pin==pin
    def get_balance(self, pin): 
        """Return account balance."""
        if self._check_pin(pin):
            return self._balance
        else:
            print("Invalid Pin")
            return -1
    def change_pin(self, oldpin, newpin): 
        """Change pin from oldpin to newpin."""
        if self._check_pin(oldpin):
            self._pin=newpin
        else:
            print("Invalid Pin")

class SavingAccount(BankAccount):
    def __init__(self,pin):
        super().__init__(pin)
        self.rate = 0.05
    def interest_add(self):
        self._balance+=self._balance*self.rate

class feeSavingAccount(SavingAccount):
    def __init__(self,pin):
        super().__init__(pin)
        self.fee = 0.5
    def withdraw(self,pin,amount):
        """Decrement account balance by amount and return amount withdrawn."""
        if self._check_pin(pin):
            if self._balance >= amount:
                self._balance-= self.fee
                self._balance-= amount
            else:
                print("Insufficient Balance")
        else:
            print("Invalid Pin")

obj1 = feeSavingAccount(121)
obj1.deposit(121,120)
obj1.get_balance(121)
obj1.withdraw(121,100)
obj1.get_balance(121)
obj1.interest_add()
obj1.get_balance(121)
```

```python
class Electronics:
    power_usage=0
    brand=''
    size=[0,0,0]
    subcategory='home'
    def __init__(self,pov,brd,size,subcat):
        self.power_usage=pov
        self.brand=brd
        self.size=size
        self.typ=subcat
    def get_brand(self):
        return self.brand()

class Microwave(Electronics):
    def __init__(self,pov,brd,size,typ,capacity,isinverter=True):
        super().__init__(pov,brd,size,typ)
        self.Capacity=capacity
        self.isinverter=isinverter
    def display_char(self):
        print(f'Brand : {self.brand}\nCapacity : {self.Capacity}\nProduct Size : {self.size}\nPower useage: {self.power_usage}\nIs Inverter ? {self.isinverter}\nSize : {self.size}\nCategory : {self.typ}')
    def get_Inverterflag(self):
        return self.isinverter
mir = Microwave(10,'LG',[2,2,2],'home','20L',False)
mir.display_char()
```
