```python
class Car:
    
    def __init__(self,name,Hp=None,F1=False,SL=False,F3=False):
        self.car_name = name
        self.ct_f1 = F1
        self.ct_f3 = F3
        self.ct_st = SL
        self.HorsePower = Hp
    
    def get_F1_certificate(self):
        return self.ct_f1
    
    def get_F3_certificate(self):
        return self.ct_f3
    
    def get_st_certificate(self):
        return self.ct_st
    
    def valid_car(self):
        return [self.ct_f1,self.ct_f3,self.ct_st].count(True)==1
    
    def __repr__(self):
        return f'{self.car_name[:3].upper()}'

cr = Car('Gladiator',100,F1=True)
cr

class Team:
 
    def __init__(self,name,TM,DR,Mec):
        self.Name = name
        self.TeamManager = TM
        self.DriverRoster = DR if type(DR)==list else [DR]
        self.Mechanics = Mec if type(Mec)==list else [Mec]
        self.cars =[]
 
    def register_car(self,name,certification):
        if certification == 'F1':
            self.cars.append(Car(name,F1=True)) 
        elif certification == 'SL':
            self.cars.append(Car(name,SL=True))
        elif certification == 'F3':
            self.cars.append(Car(name,F3=True))
    def __str__(self):
    
        return f''' 
        ######################################################################
        TEAM NAME : {self.Name}
        TEAM MANAGER : {self.TeamManager}
        DRIVER ROSTER : {self.DriverRoster}
        Mechanics : {self.Mechanics}
        CARS : {self.cars}
        ######################################################################
        '''
    
    def __repr__(self):
        return f'{self.Name}'

TeamRedBullRaching = Team("Oracle Red Bull Racing",'Christian Horner',['MAX','Sergio'],'Pierre Waché')
TeamRedBullRaching.register_car('REDBULL','F1')
TeamRedBullRaching.register_car('RLKL','F3')
print(TeamRedBullRaching)
```
