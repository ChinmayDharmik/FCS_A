# Practice Composition - Part I

```python
# Q1
class Salary:
    def __init__(self, base_pay, bonus):
        self.__base_pay = base_pay
        self.__bonus = bonus

class Employee:
    def __init__(self, name, base_pay, bonus):
        self.__name = name
        self.__salary = Salary(base_pay, bonus)
```

```python
#Q2
class Vehicle:
    def __init__(self,typ,color):
        self.type=typ
        self.color=color
    def display(self):
        print(f'type : {self.type}\ncolor : {self.color}')

class BusEngine:
    def start(self):
        print("Starting the Bus")
    def stop(self):
        print("Stopping the Bus")

class Bus(Vehicle):
    def __init__(self,typ,color):
        super().__init__(typ,color)
        self.engine = BusEngine()
    def set_start(self):
        self.engine.start()
    def set_stop(self):
        self.engine.stop()


bus1 = Bus("x1","Yellow")
bus1.display()
bus1.set_start()
bus1.set_stop()
```

# Practice Aggregation - Part II

```python
#q2
class Member:
    def __init__(self, member_id, name, society_name):
        self.__member_id = member_id
        self.__name = name
        self.__society_name = society_name
class Society:
    
    def __init__(self, society_name):
        self.__name = society_name
        self.__members = []

    def register_member(self,member_id, name):
        mem =Member(member_id,name,self.__name)
        self.__members.append(mem)
        return mem
    def get_member_count(self):
        return len(self.__members)
    

class University:
    def __init__(self, name):
        self.__name = name
        self.__societies = []
    def register_society(self,society_name):
        soc = Society(society_name)
        self.__societies.append(soc)
        return soc
    def get_society_count(self):
        return len(self.__societies)

uni = University('Adelaide')
CSclub= uni.register_society('CS Club')
AnimeCLub = uni.register_society('Anime Club')
csmem1 = CSclub.register_member(1,'Chinmay Dharmik')
csmem2 = CSclub.register_member(2,'Rohan Thomas')
csmem3 = CSclub.register_member(3,'Haoyu Chen')
Animem1 = AnimeCLub.register_member(1,'Chinmay Dharmik')
Animem2 = AnimeCLub.register_member(2,'Rohan Thomas')
Animem3 = AnimeCLub.register_member(3,'Haoyu Chen')

print(uni.get_society_count(),CSclub.get_member_count(),AnimeCLub.get_member_count())
```
