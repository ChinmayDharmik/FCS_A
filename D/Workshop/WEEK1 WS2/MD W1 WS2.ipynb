1. Write an OOP solution for the following scenario.

* Write a Rectangle class, which allows you to build a rectangle with length and width attributes.
* Create a Perimeter() method to calculate the perimeter of the rectangle and a Area() method to calculate the area of ​​the rectangle.
* Create a method display() that displays the length, width, perimeter, and area of an object created using an instantiation on rectangle class.

````python
class Rectangle:
    def __init__(self,l,w):
        self.length = l
        self.width = w

    def Perimeter(self):
        return 2*(self.length+self.width)

    def Area(self):
        return self.length*self.width
    
    def Display(self):
        print(f'length = {self.length}\nwidth = {self.width}\nPerimeter : {self.Perimeter()}\nArea : {self.Area()}')

obj  = Rectangle(5,10)
obj.Display()
````

* Define a testBelongs() method of the class, which allows testing whether a point A(x, y) belongs to the circle C(O, r) or not.

    ```python
        # Creating of an instance object

        print ("the perimeter of the circle C is:", C.perimeter() )
    ```

````python
import math 
class Circle:
    def __init__(self,a,b,r):
        self.a = a         
        self.b = b         
        self.r = r
    
    def area(self):
        return math.pi*self.r**2
    
    def perimeter(self):
        return 2*math.pi*self.r
    
    def test_belong(self,x,y):
        distance_between_points = math.sqrt((x-self.a)**2+(y-self.b)**2)
        return distance_between_points <= self.r

C = Circle (1,2,1)

print ("the perimeter of the circle C is:", C.perimeter() )
print ("the area of circle C is:", C.area())
# we test if the point(1,1) belong to the circle or not
C.test_belong(1,1)
````

```python

            self._area = area

        def __len__(self):
```

* Extend the above program to return a useful object representation (e.g., details such as area and breadth of the object) using the `__repr__` method.
* Extend the above program to get the sum of the breadth of two rectangle objects using the `__add__` method.
* Extend the above program to get the difference of areas of two rectangle objects using the `__sub__` method.

````python
class RectangleClass(object):
    def __init__(self, area, breadth):
        self._area = area
        self._breadth = breadth
        
    def __len__(self):
        return int(self._area / self._breadth)
    
    def __repr__(self):
        return f'area is: {self._area}  \nbreath is: {self._breadth}'

    
    def __add__(self,rect1):
        return rect1._breadth + self._breadth
    
    def __sub__(self,rect1):
        return rect1._area - self._area

r1 = RectangleClass(90, 5)
r2 = RectangleClass(10,2)
print(len(r1))
print(repr(r1))
print(r1 - r2)
````
