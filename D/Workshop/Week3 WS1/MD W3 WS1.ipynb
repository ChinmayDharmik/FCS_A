4. Consider you have two classes: Rectangle and Circle with the following attributes and methods.

* Rectangle Class
    * Attributes:
        * width
        * height
    * Methods:
        * get_area()
* Circle Class
    * Attributes:
        * radius
    * Methods:
        * get_area()
        
Note that in the above two classes, the method get_area() calculates the area of the corresponding shape depending on the values of their attributes. Implement the above scenario using polymorphism using methods.

```python
class rectangle:
    def __init__(self,wdt,hgt):
        self.width = wdt
        self.height = hgt
    def get_area(self):
        return self.width*self.height
import math 
class circle:
    def __init__(self,rad):
        self.radius = rad
    def get_area(self):
        return math.pi*self.radius**2

rectangle = rectangle(10,20)
circle= circle(10)

rectangle.get_area()
circle.get_area()

```

```python
class shape:
    def get_area(self):
        return 1
class rectangle(shape):
    def __init__(self,wdt,hgt):
        self.width = wdt
        self.height = hgt
    def get_area(self):
        return self.width*self.height
import math 
class circle(shape):
    def __init__(self,rad):
        self.radius = rad
    def get_area(self):
        return math.pi*self.radius**2

rectangle = rectangle(10,20)
circle= circle(10)

rectangle.get_area()
circle.get_area()
```

```python
class FullTimeEmployee:
    def __init__(self,name,position,salary):
        self.name = name
        self.position = position
        self.salary = salary
    def bonus(self):
        return 0.5*self.salary
class PartTimeEmployee(FullTimeEmployee):
    def __init__(self,name,position,salary):
        super().__init__(name,position,salary)
    def bonus(self):
        return 0.1*self.salary

emp1 = FullTimeEmployee('John', 'Frontend Developer', 10000)
print(emp1.bonus())
emp2 = PartTimeEmployee('Danise', 'Data Scientist', 20000)
print(emp2.bonus())
```

```python
class Vehicle:
    def __init__(self, fare):
        self.fare = fare
    def __add__(self,other):
        return self.fare+other.fare
    def __lt__(self,other):
        return self.fare < other.fare
        
bus= Vehicle(20)
car= Vehicle(30)

total_fare = bus + car
print(total_fare)

compare = bus < car
print(compare)
```

```python
class ComplexNumbers:
    def __init__(self,real,img):
        self.real=real
        self.img=img
    def __add__(self,other):
        return self.real+other.real,self.img+other.img  
    def __sub__(self,other):
        return self.real-other.real,self.img-other.img 

x = ComplexNumbers(2,1)
y = ComplexNumbers(3,9)
print(x+y,x-y)
```

```python
class Device:
    def __init__(self,name,price):
        self.name = name
        self.price = price

    def device_details(self):
        return f'Device Name: {self.name}\nDevice Price: {self.price}'

class Phone(Device):
    def __init__(self,name,price,color):
        super().__init__(name,price)
        self.color = color
    def device_details(self):
        return f'Phone Name: {self.name}\nPhone Price: {self.price}\nPhone color : {self.color}'
    
class Laptop(Device):
    def __init__(self,name,price,size):
        super().__init__(name,price)
        self.size = size
    def device_details(self):
        return f'Laptop Name: {self.name}\nLaptop Price: {self.price}\nLaptop size : {self.size}'

print(Laptop('lenovo',2800,16).device_details())
```
